# 필수과제 1. 파생변수 만들기
- 성별에 따라, 클래스에 따라서 값을 매핑하는 반복문 만들자
- 남, 여 pclass ,1,2,3 으로 나뉘어져 있다.

- 남 1 = m1 
- 남 2 = m2
- 남 3 = m3

- 여 1 = f1
- 여 2 = f2
- 여 3 = f3

```python
import seaborn as sns
import pandas as pd
tt = sns.load_dataset('titanic')
```

```python
tt
```

```python
tt.dtypes
```

```python
tt['pclass']
```

```python
tt['aa']= 0
```

```python
for i in range(len(tt)):
    if tt.sex.loc[i] == 'male':
        if tt.pclass.loc[i] == 1 :
            tt['aa'].loc[i] = 'm1'
        elif tt.pclass.loc[i] == 2 :
            tt['aa'].loc[i] = 'm2'
        else :
            tt['aa'].loc[i] = 'm3'
    elif tt.sex.loc[i] == 'female':
        if tt.pclass.loc[i] == 1 :
            tt['aa'].loc[i] = 'f1'
        elif tt.pclass.loc[i] == 2 :
            tt['aa'].loc[i] = 'f2'
        else :
            tt['aa'].loc[i] = 'f3'
```

```python
print(tt.aa.value_counts())
```
